```python
import sys, os
MAIN_PATH = os.path.abspath('..')
sys.path.append(MAIN_PATH)
import numpy as np
from PIL import Image

from datasets.dataset import load_dataset
```

```python
idx = 2
name_list = ['original', 'dataset_v0', 'cropped']
name  = name_list[idx]

img_list, mask_list = load_dataset(name, 
                                   data_root = None,
                                   target_long_edge = 260)
```

```python
from IPython.display import display
from PIL import Image
import numpy as np

import random

img_id = random.randint(0, len(img_list)-1)

for img_id in [img_id]:
    img = img_list[img_id]
    mask = mask_list[img_id]
    mask_uint8 = (mask.astype(np.uint8) * 255)
    mask_rgb = np.stack([mask_uint8, mask_uint8, mask_uint8], axis=-1)
    comparison = np.concatenate([img, mask_rgb], axis=1)
    display(Image.fromarray(comparison))
```
